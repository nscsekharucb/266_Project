```python
#
# Set polarity 
#
df['textblob_polarity'] = 0
df['grammaticality'] = 0
# df['semisuper_polarity'] = 0

# df.columns = ['sentences', 'textblob_polarity', 'grammaticality', 'semisuper_polarity']
```

```python
#
# Get polarity using textblob
#
def get_textblob_polarity(sentence):
    if len(sentence.split()) < 3:
        return 0
    
    blob = TextBlob(sentence)
    if blob.sentiment.subjectivity > 0.0 or blob.sentiment.polarity > 0.0 :
        return 1
    else :
        return 0

for i in range(len(df)):
    polarity = get_textblob_polarity(df.iloc[i]['sentences'])
    df.loc[i, 'textblob_polarity'] = polarity
```

```python
#
# Get grammatic check using language_check
#
import language_check

def get_grammaticality(sentence):
    if len(sentence.split()) < 3:
        return 0

    tool = language_check.LanguageTool('en-US')
    matches = tool.check(sentence)
    #
    # Did we get some errors?
    #
    if len(matches):
        return 0
    else:
        return 1

for i in range(len(df)):
    df.loc[i, 'grammaticality'] = get_grammaticality(df.iloc[i]['sentences'])
```

```python
#
# Semi supervised learning
#
from sklearn.feature_extraction.text import *
from sklearn.naive_bayes import BernoulliNB
from sklearn.grid_search import GridSearchCV
from sklearn.linear_model import LogisticRegression

# newdf = pd.read_csv('../dataset/redis_dataset.csv.annotated', encoding = "ISO-8859-1", index_col=False)
#
# Add semi supervised layer
#
newdf = df
train_sentences = newdf['sentences'][:200]
train_categories = newdf['semisuper_polarity'][:200]
test_sentences = newdf['sentences'][200:]
test_categories = newdf['semisuper_polarity'][200:]

# tf = TfidfVectorizer()
tf = CountVectorizer()
X_tf = tf.fit_transform(train_sentences)
test_tf = tf.transform(test_sentences)

#
# Bernoulli NB Model
#
alphas = {'alpha': [0.001, 0.01, 0.1, 0.5, 1.0, 2.0, 10.0]}
grid = GridSearchCV(BernoulliNB(),alphas)

grid.fit(X_tf, train_categories)
# print('Best BernoulliNB score:', grid.best_score_, 'with alpha= ', grid.best_params_)

b_clf = BernoulliNB(alpha=grid.best_params_['alpha'])
b_clf.fit(X_tf, train_categories)
test_categories = b_clf.predict(test_tf)

for i in range(200, len(newdf)):
    newdf.loc[i, 'semisuper_polarity'] = test_categories[i-200]

#
# Classify sentences as readable if atleast 2 categories are 1
#
for i in range(0, len(newdf)):
    newdf.loc[i, 'readability'] = int((newdf.loc[i, 'semisuper_polarity'] + newdf.loc[i, 'textblob_polarity'] + newdf.loc[i, 'grammaticality'] > 1))
```

```python
#
# Get sentences with consensus
# Store them in a new df
#
#
# Classify sentences as readable if atleast 2 categories are 1
#
for i in range(0, len(df)):
    df.loc[i, 'readability'] = int((df.loc[i, 'semisuper_polarity'] + df.loc[i, 'textblob_polarity'] + df.loc[i, 'grammaticality'] > 1))

df.head(25)
```
